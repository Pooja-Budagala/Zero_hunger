```python
# Define mappings for each column
data.Country.replace({'Outlier': 0,
    'India': 1,
    'USA': 2,
    'Brazil': 3,
    'China': 4,
    'Nigeria': 5,
    'Germany': 6,
    'Australia': 7,
    'Canada': 8,
    'South Africa': 9,
    'Mexico': 10
}, inplace=True)
data.Region.replace({'Outlier': 0,
    'Asia': 1,
    'North America': 2,
    'South America': 3,
    'Europe': 4,
    'Africa': 5,
    'Oceania': 6
}, inplace=True)
data.Land_Use_Type.replace({'Outlier': 0,
    'Cropland': 1,
    'Pasture': 2,
    'Residential': 3,
    'Industrial':4,
    'Deciduous Forest':5,
    'Coniferous Forest':6,
    'Arid Desert':7,
    'Semi-Arid Desert':8,
    'Wetlands':9,
    'Grasslands':10,
    'Coastal':11,
    'Urban':12
}, inplace=True)
```

```python
# Convert Agricultural_Production to categorical (if not done already)
def convert_agricultural_production(df):
    bins = [0, 300, 700, 1000]
    labels = ['Low', 'Medium', 'High']
    df['Agricultural_Production_Category'] = pd.cut(df['Agricultural_Production'], bins=bins, labels=labels)
    return df

# Apply the conversion
data = convert_agricultural_production(data)
data.head()
```

```python
# Drop the original Agricultural_Production column
data = data.drop(columns=['Agricultural_Production'])
```

```python
data.nunique()
```

```python
#Dropping unwanted data because of having more unique values
data.drop(['Population'], axis=1, inplace=True)
data.drop(['Food_Insecurity_Rate'], axis=1, inplace=True)
data.drop(['Access_to_Clean_Water'], axis=1, inplace=True)
data.drop(['Soil_Moisture'], axis=1, inplace=True)
data.drop(['Crop_Health_Index'], axis=1, inplace=True)
data.drop(['Pest_Infestation_Level'], axis=1, inplace=True)
data.drop(['Water_Availability'], axis=1, inplace=True)
data.drop(['IoT_Sensor_Data'], axis=1, inplace=True)
data.drop(['Market_Prices'], axis=1, inplace=True)
```

```python
# Handling outliers by replacing with median
median_premium = data["GDP_Per_Capita"].median()
data["GDP_Per_Capita"] = data["GDP_Per_Capita"].apply(lambda x: median_premium if x > 1.5 * median_premium else x)
```

```python
median_premium = data["Year"].median()
data["Year"] = data["Year"].apply(lambda x: median_premium if x > 1.5 * median_premium else x)
```

```python
median_premium = data["Month"].median()
data["Month"] = data["Month"].apply(lambda x: median_premium if x > 1.5 * median_premium else x)
```

```python
median_premium = data["Unemployment_Rate"].median()
data["Unemployment_Rate"] = data["Unemployment_Rate"].apply(lambda x: median_premium if x > 1.5 * median_premium else x)
```

```python
median_premium = data["Average_Temperature"].median()
data["Average_Temperature"] = data["Average_Temperature"].apply(lambda x: median_premium if x > 1.5 * median_premium else x)
```

```python
median_premium = data["Rainfall"].median()
data["Rainfall"] = data["Rainfall"].apply(lambda x: median_premium if x > 1.5 * median_premium else x)
```

```python
median_premium = data["Disease_Outbreaks"].median()
data["Disease_Outbreaks"] = data["Disease_Outbreaks"].apply(lambda x: median_premium if x > 1.5 * median_premium else x)
```

```python
def evaluate_model(act, pred):
    from sklearn.metrics import confusion_matrix,classification_report, accuracy_score, recall_score, precision_score, f1_score
    print("Confusion Matrix \n", confusion_matrix(act, pred))
    print(classification_report(act,pred))
    print("Accurcay : ", accuracy_score(act, pred))
    print("Recall   : ", recall_score(act, pred,average='weighted'))
    print("Precision: ", precision_score(act, pred, average='weighted'))
    print("F1_score : ", f1_score(act, pred, average='weighted'))
```
